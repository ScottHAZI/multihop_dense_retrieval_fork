```python
import json

def get_file_num_lines(path):
    i = 0
    with open(path, 'r') as f:
        for line in f:
            i += 1
    return i

def read_file_line_as_json(path, num_lines=100):
    # num_lines = -1 if all lines are requried
    json_data = []
    i = 0
    with open(path, 'r') as f:
        for line in f:
            i += 1
            json_data.append(json.loads(line))
            if i == num_lines:
                break
    return json_data

def print_file_line(path, num_lines=3):
    i = 0
    print(f"{path}:\n")
    # num_lines = -1 if all lines are requried
    with open(path, 'r') as f:
        for line in f:
            i += 1
            print(line)
            if i == num_lines:
                break
    print('')
    
def print_json_content(obj):
    for key, value in obj.items():
        if isinstance(value, list):
            print(key, '[list]:')
            for idx, sub_value in enumerate(value):
                print(f"{idx}. {sub_value}")
            print()
                
        else:
            print(f"{key}:\n{value}\n")

hotpot_train_with_neg_v0_file_path     = '../data/hotpot/hotpot_train_with_neg_v0.json'
hotpot_dev_with_neg_v0_file_path       = '../data/hotpot/hotpot_dev_with_neg_v0.json'
hotpot_qas_val_file_path               = '../data/hotpot/hotpot_qas_val.json'
dev_retrieval_b50_k50_sp_file_path     = '../data/hotpot/dev_retrieval_b50_k50_sp.json'
dev_retrieval_top100_sp_file_path      = '../data/hotpot/dev_retrieval_top100_sp.json'
train_retrieval_b100_k100_sp_file_path = '../data/hotpot/train_retrieval_b100_k100_sp.json'
```

```python
hotpot_train_with_neg_v0_json     = read_file_line_as_json(hotpot_train_with_neg_v0_file_path)
hotpot_dev_with_neg_v0_json       = read_file_line_as_json(hotpot_dev_with_neg_v0_file_path)
hotpot_qas_val_json               = read_file_line_as_json(hotpot_qas_val_file_path)
dev_retrieval_b50_k50_sp_json     = read_file_line_as_json(dev_retrieval_b50_k50_sp_file_path)
dev_retrieval_top100_sp_json      = read_file_line_as_json(dev_retrieval_top100_sp_file_path)
train_retrieval_b100_k100_sp_json = read_file_line_as_json(train_retrieval_b100_k100_sp_file_path)
```

```python
print(len(hotpot_train_with_neg_v0_json), '\n')
print(hotpot_train_with_neg_v0_json[0].keys(), '\n')
print_json_content(hotpot_train_with_neg_v0_json[0])
```

```python
print(len(hotpot_dev_with_neg_v0_json), '\n')
print(hotpot_dev_with_neg_v0_json[0].keys(), '\n')
print_json_content(hotpot_dev_with_neg_v0_json[0])
```

```python
print(len(hotpot_qas_val_json), '\n')
print(hotpot_qas_val_json[0].keys(), '\n')
print_json_content(hotpot_qas_val_json[0])
```

```python
import wikipediaapi

wiki_wiki = wikipediaapi.Wikipedia('Multi-hop QA (jinghang@vibe.us.com)', 'en')
```

```python
num_page_found = 0
num_page_missing = 0
num = 0
for k, v in wiki_id2doc_json.items():
    num += 1
    page = wiki_wiki.page(v['title'])
    if page.exists():
        num_page_found += 1
    else:
        num_page_missing += 1
    if num == 10000:
        break
print(num_page_found, num_page_missing)
```
